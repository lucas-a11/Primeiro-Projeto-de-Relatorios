```python
import pandas as pd 
tabela = pd.read_excel ("Vendas.xlsx")
tabela = tabela.drop (columns="Código Venda")
display (tabela)
```

```python
tabela = tabela.dropna()
display (tabela.info())

```

```python
faturamento = tabela[["ID Loja","Valor Final"]].groupby("ID Loja").sum()
display (faturamento)
```

```python
qtd_vnd = tabela[["Quantidade","Produto"]].groupby("Produto").sum()
display (qtd_vnd)
```

```python
data = tabela[["Data","Quantidade"]].groupby("Data").sum()
display (data)
```

```python
meta_faturamento = 1700000  # Defina a meta de faturamento aqui

# Verifique se o faturamento de cada loja atingiu a meta
for i, valor_faturamento in faturamento.iterrows():
    if valor_faturamento['Valor Final'] >= meta_faturamento:
        print(f"A loja {i} atingiu a meta de faturamento!")
    else:
        print(f"A loja {i} não atingiu a meta de faturamento.")
```

```python
qtd_loja = tabela[["ID Loja", "Quantidade"]].groupby("ID Loja").sum()
display (qtd_loja)
```
